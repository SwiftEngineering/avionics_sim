```python
# Imports
import math
import numpy as np
from scipy.signal import lti
from scipy.interpolate import interp2d
import matplotlib.pyplot as plt
```

For the most simplistic of disturbances we can apply a gaussian distirbuted noise directly to a strength and direction vector to determine the incident linear rate onto an aircraft.

## Given
Desired turbulence:
- strength
- variance
- direction 
- variance of the direction

## Find

- linear rate of turbulence

## Assumption
It will be assumed that a random noise that follows a gaussian distribution represents the turbulence.

## Analysis

```python
# Given

strength_intensity_m_per_s = 10
strength_variance = 0.01
direction_intensity = 1
direction_variance = 0.01
```

```python
# Assumption

sim_start_s = 0
sim_end_s = 10
dt_s = 0.01

t_span = [sim_start_s, sim_end_s]
t_s = np.arange(t_span[0], t_span[1] + dt_s/2, dt_s)

noise_strength = np.random.normal(strength_intensity_m_per_s, np.sqrt(strength_variance), t_s.shape[0]) 
noise_x = np.random.normal(direction_intensity, np.sqrt(direction_variance), t_s.shape[0])
noise_y = np.random.normal(direction_intensity, np.sqrt(direction_variance), t_s.shape[0])
noise_z = np.random.normal(direction_intensity, np.sqrt(direction_variance), t_s.shape[0])

```

```python
linear_rate_m_per_s = []

for step in range(0, len(t_s)):
    strength_m_per_s = noise_strength[step]
    direction = np.array([noise_x[step], noise_y[step], noise_z[step]])
    direction = direction / np.linalg.norm(direction)
    linear_rate_m_per_s.append(strength_m_per_s * direction)

linear_rate_m_per_s = np.array(linear_rate_m_per_s);

print(linear_rate_m_per_s)
```

```python
plt.plot(t_s, linear_rate_m_per_s[:,0])
```

```python
plt.plot(t_s, linear_rate_m_per_s[:,1])
```

```python
plt.plot(t_s, linear_rate_m_per_s[:,2])
```

## Conclusion

While not a very elegant solution, the turbulence is effictively gaussian noise, this does provide some initial turbulent behavior. Again the use case is for moments when the aircraft is in hover, or sufficiently slow, which coincides with the more costly behavior of the aircraft. Furthermore, even this simplisitic model seems to be shared not just by one, but by two other UAV simulator teams in the form of the ![PX4-Team][https://github.com/PX4/PX4-SITL_gazebo/blob/master/src/gazebo_wind_plugin.cpp] and ![ETHZ-ASL-Team][https://github.com/ethz-asl/rotors_simulator/blob/master/rotors_gazebo_plugins/src/gazebo_wind_plugin.cpp] Their approach varies in that they directly apply the wind as a force where the wind generated here is fed to the aerodynamic model.
