# Wind Modeling

The objective of wind modeling is to ascertain the world linear and angular wind rates. These wind rates will be provided to the aerodynamic model in conjunction with airfoil velocity to provide the incident freestreem velocity.

In general sense world wind will be represented as:

$$ Wind\ Rate(world)\ =\ Sustained\ Wind\ +\ Turbulent\ Wind$$

The matter of a sustained wind is a trivial matter; namely sustained wind is a constant linear rate vector for the duration of flight. 

Of more consequence is the matter of modeling turbulent wind.

## Turbulent Wind

The FAA has outlined the Von Karman as the preffered model for generating turbulent winds; unfortunately, it and other preferred models all suffer from the common problem of representing turbulence as a frozen field. That makes them non-ideal for all situations where an aircraft is moving sufficeintly slow as is a common occurance for a tailsitter aircraft that spends some of its time in hover.

The idea then is to use one model to represent turbulence at zero-low aircraft velocities and another at velocities sufficient for a frozen field model. That is:

$$ Turbulent\ Wind = \begin{cases} gaussian\ wind & \text{if } V < V_{threshold} \\ dryden/von-karman\ wind, & \text{otherwise} \end{cases} $$

The use of Dryden over Von Karman comes down to the required fidelity of the model at the cost of computational complexity. The Von Karman while considered more accuracte introduces computational overhead in managing the possible unreal values from the square root of the PSD used in the formulation of the transfer function.

## Wind Notation (FAA/METAR)

It can be seen from the FAA, that in Chapter 13, p13-6 of the [Pilots Handbook of Aeronautical Knowledge](https://www.faa.gov/regulations_policies/handbooks_manuals/aviation/phak/media/15_phak_ch13.pdf) that wind shall be designated by the form:

For the cases of the turbulence models that utilize a mean and variance to the gust the following has been the interpretation of the FAA standards.

Given a wind defintion of the form:
$$ \pmb{ddd} \pmb{ss(s)} \mathrm{G} \pmb{tt(t)} kts$$
where
$$ \pmb{ddd} = \textrm{direction in 10s of degrees the wind is blowing from, clockwise from true north}$$
$$ \pmb{ss(s)} = \textrm{sustained wind strength}$$
$$ \pmb{tt(t)} = \textrm{max instaneuous wind speed as recorded in a 10 minute interval}$$ 

note: dy defintion, the gusting value $\pmb{tt(t)}$ must always be greater than \pmb{ss(s)}

By the FAA definition a wind defined as $09020G35Kts$, should resolve to:
- Wind coming from 90 degrees off North (East), moving from East to West
- Wind has a minimum strength of 20kts
- Wind has max gust of 35kts

What we can infer from this defintion is that there is a constant or 'sustained wind' of 20kts at all times. Furthermore, we can add a 'turbulent wind' that is of random form(stochastic process) within the range defined as:

$$ [0, G_{max}] $$

where,

$$ G_{max} = \pmb{tt(t)} - \pmb{ss(s)} $$ 

This turbulent wind can then be said to have mean strength, offset from the sustained strength, of:
$$ \mu_{turbulence} = \frac{G_{max}}{2}$$
and defining the bounds within 3 standard deviations of the mean, then it follows that the standard deviation is:
$$ \sigma_{turbulence} = \frac{\frac{G_{max}}{2}}{3} = \frac{G_{max}}{6} $$
and the variance:
$$ {\sigma_{turbulence}}^2 = \frac{G_{max}^2}{36} $$

thus, for our original wind defintion of $09020G35Kts$, the turbulent wind can be said to have:
- mean wind strength, $\mu_{turbulence}$, of $7.5kts$
- wind variance of $6.25kts^2$
